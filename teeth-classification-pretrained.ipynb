```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import warnings
warnings.filterwarnings('ignore')

# --------------------------
# 🔧 1. تحديد مسارات المجلدات
# --------------------------
base_dir = '/kaggle/input/teeth-dataset-for-classification/Teeth_Dataset'
train_dir = os.path.join(base_dir, 'Training')
val_dir = os.path.join(base_dir, 'Validation')
test_dir = os.path.join(base_dir, 'Testing')

# تأكد من أن المجلدات موجودة
for path in [train_dir, val_dir, test_dir]:
    if not os.path.exists(path):
        raise FileNotFoundError(f"المجلد غير موجود: {path}")

# --------------------------
# 🛠️ 2. Preprocessing: ImageDataGenerator مع Data Augmentation
# --------------------------
img_size = 224
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    shear_range=0.2,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# تحميل البيانات
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False  # مهم للتصنيف
)

# طباعة تفاصيل الفئات
print("عدد الفئات:", train_generator.num_classes)
print("أسماء الفئات:", list(train_generator.class_indices.keys()))
print("عدد صور التدريب:", train_generator.samples)
print("عدد صور التحقق:", val_generator.samples)
print("عدد صور الاختبار:", test_generator.samples)

# --------------------------
# 🤖 3. استخدام Pre-trained Model (ResNet50)
# --------------------------
base_model = ResNet50(
    weights='imagenet',
    include_top=False,
    input_shape=(img_size, img_size, 3)
)

# تجميد الطبقات الأساسية
base_model.trainable = False

# إضافة الطبقات المخصصة
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
predictions = Dense(train_generator.num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

# --------------------------
# ⚙️ 4. تهيئة التدريب
# --------------------------
model.compile(
    optimizer=Adam(learning_rate=0.0001),  # معدل تعلم أقل
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# --------------------------
# 📊 5. حساب أوزان الفئات (للتوازن)
# --------------------------
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes
)
class_weight_dict = dict(enumerate(class_weights))

# --------------------------
# 🛑 6. Callbacks: التوقف المبكر وتقليل معدل التعلم
# --------------------------
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-7)

callbacks = [early_stopping, reduce_lr]

# --------------------------
# 🚀 7. تدريب النموذج
# --------------------------
print("\n[+] بدء التدريب...")
history = model.fit(
    train_generator,
    epochs=20,
    validation_data=val_generator,
    class_weight=class_weight_dict,
    callbacks=callbacks
)

# --------------------------
# 📈 8. تقييم النموذج على مجموعة الاختبار
# --------------------------
print("\n[+] تقييم النموذج...")
test_loss, test_acc = model.evaluate(test_generator, verbose=1)
print(f"\n✅ دقة الاختبار: {test_acc:.4f}")

# التنبؤ بالنتائج
y_pred = model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes

# التقرير
class_names = list(test_generator.class_indices.keys())
print("\n📋 تقرير التصنيف:")
print(classification_report(y_true, y_pred_classes, target_names=class_names))

# --------------------------
# 🎨 9. Visualization: منحنيات الدقة والخسارة
# --------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(history.history['accuracy'], label='دقة التدريب')
ax1.plot(history.history['val_accuracy'], label='دقة التحقق')
ax1.set_title('دقة النموذج')
ax1.set_xlabel('الدورة (Epoch)')
ax1.set_ylabel('الدقة')
ax1.legend()

ax2.plot(history.history['loss'], label='الخسارة في التدريب')
ax2.plot(history.history['val_loss'], label='الخسارة في التحقق')
ax2.set_title('خسارة النموذج')
ax2.set_xlabel('الدورة (Epoch)')
ax2.set_ylabel('الخسارة')
ax2.legend()

plt.tight_layout()
plt.show()

# --------------------------
# 🔺 10. مصفوفة الالتباس
# --------------------------
cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('مصفوفة الالتباس')
plt.xlabel('التصنيف المتوقع')
plt.ylabel('التصنيف الفعلي')
plt.show()

# --------------------------
# 🖼️ 11. عرض صور مع التنبؤ (مُصلح)
# --------------------------
def plot_predictions():
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))
    axes = axes.ravel()

    for i in range(6):
        idx = np.random.randint(0, len(test_generator))
        img, label = test_generator[idx]
        
        # التنبؤ باستخدام النموذج
        pred = model.predict(img, verbose=0)
        pred_class = np.argmax(pred[0])

        # استخراج التصنيف الحقيقي
        true_class_index = np.argmax(label[0]) if len(label[0]) > 1 else int(label[0])
        true_class_name = class_names[true_class_index]
        pred_class_name = class_names[pred_class]

        # عرض الصورة
        axes[i].imshow(img[0])
        axes[i].set_title(f"الحقيقي: {true_class_name}\nالتنبؤ: {pred_class_name}", fontsize=10)
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()

print("\n[+] عرض بعض التنبؤات:")
plot_predictions()

# --------------------------
# 💾 12. حفظ النموذج (اختياري)
# --------------------------
model.save('teeth_classification_model.h5')
print("\n✅ تم حفظ النموذج في 'teeth_classification_model.h5'")
```
