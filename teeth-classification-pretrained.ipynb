```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import warnings
warnings.filterwarnings('ignore')

# --------------------------
# ğŸ”§ 1. ØªØ­Ø¯ÙŠØ¯ Ù…Ø³Ø§Ø±Ø§Øª Ø§Ù„Ù…Ø¬Ù„Ø¯Ø§Øª
# --------------------------
base_dir = '/kaggle/input/teeth-dataset-for-classification/Teeth_Dataset'
train_dir = os.path.join(base_dir, 'Training')
val_dir = os.path.join(base_dir, 'Validation')
test_dir = os.path.join(base_dir, 'Testing')

# ØªØ£ÙƒØ¯ Ù…Ù† Ø£Ù† Ø§Ù„Ù…Ø¬Ù„Ø¯Ø§Øª Ù…ÙˆØ¬ÙˆØ¯Ø©
for path in [train_dir, val_dir, test_dir]:
    if not os.path.exists(path):
        raise FileNotFoundError(f"Ø§Ù„Ù…Ø¬Ù„Ø¯ ØºÙŠØ± Ù…ÙˆØ¬ÙˆØ¯: {path}")

# --------------------------
# ğŸ› ï¸ 2. Preprocessing: ImageDataGenerator Ù…Ø¹ Data Augmentation
# --------------------------
img_size = 224
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    shear_range=0.2,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_size, img_size),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False  # Ù…Ù‡Ù… Ù„Ù„ØªØµÙ†ÙŠÙ
)

# Ø·Ø¨Ø§Ø¹Ø© ØªÙØ§ØµÙŠÙ„ Ø§Ù„ÙØ¦Ø§Øª
print("Ø¹Ø¯Ø¯ Ø§Ù„ÙØ¦Ø§Øª:", train_generator.num_classes)
print("Ø£Ø³Ù…Ø§Ø¡ Ø§Ù„ÙØ¦Ø§Øª:", list(train_generator.class_indices.keys()))
print("Ø¹Ø¯Ø¯ ØµÙˆØ± Ø§Ù„ØªØ¯Ø±ÙŠØ¨:", train_generator.samples)
print("Ø¹Ø¯Ø¯ ØµÙˆØ± Ø§Ù„ØªØ­Ù‚Ù‚:", val_generator.samples)
print("Ø¹Ø¯Ø¯ ØµÙˆØ± Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±:", test_generator.samples)

# --------------------------
# ğŸ¤– 3. Ø§Ø³ØªØ®Ø¯Ø§Ù… Pre-trained Model (ResNet50)
# --------------------------
base_model = ResNet50(
    weights='imagenet',
    include_top=False,
    input_shape=(img_size, img_size, 3)
)

# ØªØ¬Ù…ÙŠØ¯ Ø§Ù„Ø·Ø¨Ù‚Ø§Øª Ø§Ù„Ø£Ø³Ø§Ø³ÙŠØ©
base_model.trainable = False

# Ø¥Ø¶Ø§ÙØ© Ø§Ù„Ø·Ø¨Ù‚Ø§Øª Ø§Ù„Ù…Ø®ØµØµØ©
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.3)(x)
predictions = Dense(train_generator.num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

# --------------------------
# âš™ï¸ 4. ØªÙ‡ÙŠØ¦Ø© Ø§Ù„ØªØ¯Ø±ÙŠØ¨
# --------------------------
model.compile(
    optimizer=Adam(learning_rate=0.0001),  # Ù…Ø¹Ø¯Ù„ ØªØ¹Ù„Ù… Ø£Ù‚Ù„
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# --------------------------
# ğŸ“Š 5. Ø­Ø³Ø§Ø¨ Ø£ÙˆØ²Ø§Ù† Ø§Ù„ÙØ¦Ø§Øª (Ù„Ù„ØªÙˆØ§Ø²Ù†)
# --------------------------
class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes
)
class_weight_dict = dict(enumerate(class_weights))

# --------------------------
# ğŸ›‘ 6. Callbacks: Ø§Ù„ØªÙˆÙ‚Ù Ø§Ù„Ù…Ø¨ÙƒØ± ÙˆØªÙ‚Ù„ÙŠÙ„ Ù…Ø¹Ø¯Ù„ Ø§Ù„ØªØ¹Ù„Ù…
# --------------------------
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-7)

callbacks = [early_stopping, reduce_lr]

# --------------------------
# ğŸš€ 7. ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
# --------------------------
print("\n[+] Ø¨Ø¯Ø¡ Ø§Ù„ØªØ¯Ø±ÙŠØ¨...")
history = model.fit(
    train_generator,
    epochs=20,
    validation_data=val_generator,
    class_weight=class_weight_dict,
    callbacks=callbacks
)

# --------------------------
# ğŸ“ˆ 8. ØªÙ‚ÙŠÙŠÙ… Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¹Ù„Ù‰ Ù…Ø¬Ù…ÙˆØ¹Ø© Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±
# --------------------------
print("\n[+] ØªÙ‚ÙŠÙŠÙ… Ø§Ù„Ù†Ù…ÙˆØ°Ø¬...")
test_loss, test_acc = model.evaluate(test_generator, verbose=1)
print(f"\nâœ… Ø¯Ù‚Ø© Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±: {test_acc:.4f}")

# Ø§Ù„ØªÙ†Ø¨Ø¤ Ø¨Ø§Ù„Ù†ØªØ§Ø¦Ø¬
y_pred = model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes

# Ø§Ù„ØªÙ‚Ø±ÙŠØ±
class_names = list(test_generator.class_indices.keys())
print("\nğŸ“‹ ØªÙ‚Ø±ÙŠØ± Ø§Ù„ØªØµÙ†ÙŠÙ:")
print(classification_report(y_true, y_pred_classes, target_names=class_names))

# --------------------------
# ğŸ¨ 9. Visualization: Ù…Ù†Ø­Ù†ÙŠØ§Øª Ø§Ù„Ø¯Ù‚Ø© ÙˆØ§Ù„Ø®Ø³Ø§Ø±Ø©
# --------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(history.history['accuracy'], label='Ø¯Ù‚Ø© Ø§Ù„ØªØ¯Ø±ÙŠØ¨')
ax1.plot(history.history['val_accuracy'], label='Ø¯Ù‚Ø© Ø§Ù„ØªØ­Ù‚Ù‚')
ax1.set_title('Ø¯Ù‚Ø© Ø§Ù„Ù†Ù…ÙˆØ°Ø¬')
ax1.set_xlabel('Ø§Ù„Ø¯ÙˆØ±Ø© (Epoch)')
ax1.set_ylabel('Ø§Ù„Ø¯Ù‚Ø©')
ax1.legend()

ax2.plot(history.history['loss'], label='Ø§Ù„Ø®Ø³Ø§Ø±Ø© ÙÙŠ Ø§Ù„ØªØ¯Ø±ÙŠØ¨')
ax2.plot(history.history['val_loss'], label='Ø§Ù„Ø®Ø³Ø§Ø±Ø© ÙÙŠ Ø§Ù„ØªØ­Ù‚Ù‚')
ax2.set_title('Ø®Ø³Ø§Ø±Ø© Ø§Ù„Ù†Ù…ÙˆØ°Ø¬')
ax2.set_xlabel('Ø§Ù„Ø¯ÙˆØ±Ø© (Epoch)')
ax2.set_ylabel('Ø§Ù„Ø®Ø³Ø§Ø±Ø©')
ax2.legend()

plt.tight_layout()
plt.show()

# --------------------------
# ğŸ”º 10. Ù…ØµÙÙˆÙØ© Ø§Ù„Ø§Ù„ØªØ¨Ø§Ø³
# --------------------------
cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Ù…ØµÙÙˆÙØ© Ø§Ù„Ø§Ù„ØªØ¨Ø§Ø³')
plt.xlabel('Ø§Ù„ØªØµÙ†ÙŠÙ Ø§Ù„Ù…ØªÙˆÙ‚Ø¹')
plt.ylabel('Ø§Ù„ØªØµÙ†ÙŠÙ Ø§Ù„ÙØ¹Ù„ÙŠ')
plt.show()

# --------------------------
# ğŸ–¼ï¸ 11. Ø¹Ø±Ø¶ ØµÙˆØ± Ù…Ø¹ Ø§Ù„ØªÙ†Ø¨Ø¤ (Ù…ÙØµÙ„Ø­)
# --------------------------
def plot_predictions():
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))
    axes = axes.ravel()

    for i in range(6):
        idx = np.random.randint(0, len(test_generator))
        img, label = test_generator[idx]
        
        # Ø§Ù„ØªÙ†Ø¨Ø¤ Ø¨Ø§Ø³ØªØ®Ø¯Ø§Ù… Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
        pred = model.predict(img, verbose=0)
        pred_class = np.argmax(pred[0])

        # Ø§Ø³ØªØ®Ø±Ø§Ø¬ Ø§Ù„ØªØµÙ†ÙŠÙ Ø§Ù„Ø­Ù‚ÙŠÙ‚ÙŠ
        true_class_index = np.argmax(label[0]) if len(label[0]) > 1 else int(label[0])
        true_class_name = class_names[true_class_index]
        pred_class_name = class_names[pred_class]

        # Ø¹Ø±Ø¶ Ø§Ù„ØµÙˆØ±Ø©
        axes[i].imshow(img[0])
        axes[i].set_title(f"Ø§Ù„Ø­Ù‚ÙŠÙ‚ÙŠ: {true_class_name}\nØ§Ù„ØªÙ†Ø¨Ø¤: {pred_class_name}", fontsize=10)
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()

print("\n[+] Ø¹Ø±Ø¶ Ø¨Ø¹Ø¶ Ø§Ù„ØªÙ†Ø¨Ø¤Ø§Øª:")
plot_predictions()

# --------------------------
# ğŸ’¾ 12. Ø­ÙØ¸ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ (Ø§Ø®ØªÙŠØ§Ø±ÙŠ)
# --------------------------
model.save('teeth_classification_model.h5')
print("\nâœ… ØªÙ… Ø­ÙØ¸ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ ÙÙŠ 'teeth_classification_model.h5'")
```
